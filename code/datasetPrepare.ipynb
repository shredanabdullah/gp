```python
import os
import zipfile

def extract_zipped_folders(directory):
    if not os.path.exists(directory):
        print(f"The directory {directory} does not exist.")
        return
    files = os.listdir(directory)
    for file in files:
        file_path = os.path.join(directory, file)
        if zipfile.is_zipfile(file_path):
            extract_dir = os.path.join(directory, file[:-4])
            os.makedirs(extract_dir, exist_ok=True)
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                zip_ref.extractall(extract_dir)
                print(f"Extracted {file} into {extract_dir}")
if __name__ == "__main__":
    path_to_directory = r'C:\Users\Tobi\Desktop\pets'
    extract_zipped_folders(path_to_directory)
```

```python
def delete_zipped_folders(directory):
    if not os.path.exists(directory):
        print(f"The directory {directory} does not exist.")
        return
    files = os.listdir(directory)
    for file in files:
        file_path = os.path.join(directory, file)
        if file_path.endswith('.zip'):
            try:
                os.remove(file_path)
                print(f"Deleted {file_path}")
            except Exception as e:
                print(f"Failed to delete {file_path}: {e}")

if __name__ == "__main__":
    path_to_directory = r'C:\Users\Tobi\Desktop\charity'
    delete_zipped_folders(path_to_directory)
```

```python
import shutil
def move_inner_folders_to_main(main_directory):
    if not os.path.exists(main_directory):
        print(f"The directory {main_directory} does not exist.")
        return
    for folder in os.listdir(main_directory):
        folder_path = os.path.join(main_directory, folder)
        
        if os.path.isdir(folder_path):
            for subfolder in os.listdir(folder_path):
                subfolder_path = os.path.join(folder_path, subfolder)
                if os.path.isdir(subfolder_path):
                    destination_path = os.path.join(main_directory, subfolder)
                    try:
                        shutil.move(subfolder_path, destination_path)
                        print(f"Moved {subfolder_path} to {destination_path}")
                    except Exception as e:
                        print(f"Failed to move {subfolder_path} to {destination_path}: {e}")

if __name__ == "__main__":
    main_directory = r'C:\Users\Tobi\Desktop\pets'
    move_inner_folders_to_main(main_directory)
```

```python
def move_and_delete_intermediate_folders(main_directory):
    if not os.path.exists(main_directory):
        print(f"The directory {main_directory} does not exist.")
        return
    for folder in os.listdir(main_directory):
        folder_path = os.path.join(main_directory, folder)
        if os.path.isdir(folder_path):
            for subfolder in os.listdir(folder_path):
                subfolder_path = os.path.join(folder_path, subfolder)
                if os.path.isdir(subfolder_path):
                    destination_path = os.path.join(main_directory, subfolder)
                    try:
                        shutil.move(subfolder_path, destination_path)
                        print(f"Moved {subfolder_path} to {destination_path}")
                    except Exception as e:
                        print(f"Failed to move {subfolder_path} to {destination_path}: {e}")
            try:
                os.rmdir(folder_path)
                print(f"Deleted intermediate folder {folder_path}")
            except Exception as e:
                print(f"Failed to delete intermediate folder {folder_path}: {e}")

if __name__ == "__main__":
    main_directory = r'C:\Users\Tobi\Desktop\fashion'
    move_and_delete_intermediate_folders(main_directory)
```
